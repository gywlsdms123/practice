```python
import PIL.Image as Image
import numpy as np
```

```python
im = Image.open('D:/dataset2/femur/78post.jpg')

# viewer로 띄우기 (pillow)
#im.show() 

print(im.size)

# Fetch image pixel data to numpy array
pixImg = np.array(im)
print(pixImg.shape)
```

```python
import matplotlib.pyplot as plt

plt.imshow(pixImg)
```

```python
plt.imshow(pixImg, cmap='gray')
```

```python
# 이미지 Numpy 객체의 특정 픽셀을 가리킵니다.
px = pixImg[450, 2500]

# B, G, R 순서로 출력됩니다.
# (단, Gray Scale인 경우에는 B, G, R로 구분되지 않습니다.)
print(px)
```

```python
pixImg1 = pixImg

print(pixImg1)

mask = pixImg1 < 200
print(mask)

pixImg1[mask] = 0
print(pixImg1)
```

```python
plt.imshow(pixImg1)
```

```python
# 저장
plt.imsave('mask.jpg', pixImg1)
```
